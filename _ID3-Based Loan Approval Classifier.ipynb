```python
import math
import pandas as pd  
```

```python
data = pd.DataFrame({
    "CustomerID": ["C1", "C2", "C3", "C4", "C5"],
    "CreditScore": ["High", "Low", "Medium", "Low", "High"],
    "IncomeLevel": ["High", "Medium", "High", "Low", "Medium"],
    "LoanAmount": ["Low", "High", "Medium", "Medium", "Low"],
    "ExistingLoan": ["No", "Yes", "No", "Yes", "No"],
    "Approved": ["Yes", "No", "Yes", "No", "Yes"]
})
```

```python
def entropy(target_col):
    elements, counts = np.unique(target_col, return_counts=True)
    entropy = 0
    for i in range(len(elements)):
        probability = counts[i] / np.sum(counts)
        entropy -= probability * math.log2(probability)
    return entropy

# Function to compute Information Gain
def information_gain(data, split_attribute_name, target_name="Approved"):
    # Total entropy
    total_entropy = entropy(data[target_name])
    
    # Weighted Entropy
    vals, counts = np.unique(data[split_attribute_name], return_counts=True)
    weighted_entropy = 0
    for i in range(len(vals)):
        subset = data[data[split_attribute_name] == vals[i]]
        weighted_entropy += (counts[i] / np.sum(counts)) * entropy(subset[target_name])
    
    info_gain = total_entropy - weighted_entropy
    return info_gain
```

```python
# Calculate target entropy
target_entropy = entropy(data["Approved"])
print(f"Target Entropy: {target_entropy:.4f}")

# Compute information gain for each attribute
attributes = ["CreditScore", "IncomeLevel", "LoanAmount", "ExistingLoan"]
for attr in attributes:
    ig = information_gain(data, attr)
    print(f"Information Gain for {attr}: {ig:.4f}")
```

```python
from sklearn.tree import DecisionTreeClassifier, export_text

# Encode categorical variables
encoded_data = pd.get_dummies(data[attributes])

# Encode target
target = data["Approved"].map({"Yes":1, "No":0})

# Fit decision tree
clf = DecisionTreeClassifier(criterion="entropy")
clf.fit(encoded_data, target)

# Print tree
tree_rules = export_text(clf, feature_names=list(encoded_data.columns))
print("\nDecision Tree:\n")
print(tree_rules)
```
